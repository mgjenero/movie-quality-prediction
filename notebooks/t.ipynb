```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# Load data
data_path = "../data/processed/movie_metadata_processed.csv"
df = pd.read_csv(data_path, encoding='latin1')

X = df.drop(columns=['is_good'])
y = df['is_good']

# Store accuracies
accuracies = []

# Repeat 10 random runs
for i in range(10):
    # Split into train/test with a different random state each run
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=i
    )
    
    # Initialize and train Random Forest
    rf = RandomForestClassifier(
        n_estimators=100,
        max_depth=None,
        n_jobs=-1,
        random_state=i
    )
    rf.fit(X_train, y_train)
    
    # Predict and compute accuracy
    y_pred = rf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    print(f"Run {i+1}: Accuracy = {acc:.4f}")

# Summary
print(f"\nMean Accuracy over 10 runs: {np.mean(accuracies):.4f}")
print(f"Standard Deviation: {np.std(accuracies):.4f}")
```
